Для решения поставленной задачи вам пригодится умение работать с файлами. На данном уроке вам даны текстовые файлы, содержание числовые данные. Для того, чтобы прочитать весь файл целиком, преобразовать данные в числовой вид и разместить их в списке можно использовать следующий исходный код:

```python
f = open('1_1.txt', 'r')
data = list(map(int, f.read().split()))
f.close()
```

После выполнения указанного фрагмента исходного кода в списке data будут содержаться все числа, прочитанные из файла f. Для того, чтобы убедиться в правильности указанного кода можно вывести часть содержимого списка data и его размер.

```python
print(data[0], data[1])
print(len(data))
```

Для решения поставленной в первом уроке задания вам подадобится билиотека matplotlib. Официальную документацию для данной библиотеки можно найти [на сайте matplotlib.org](https://matplotlib.org/). Следует также отметить, что для данной библиотеки имеется большое количество примеров в интернете.

Основное назначение библиотеки matplotlib - визуализация данных в различных форматах. Например для визуализации точки можно использовать фрагмент исходного кода:

```python
import matplotlib.pyplot as plt
plt.scatter(2, 3)
plt.show()
```

В случае необходимости визуализации нескольких точек можно делать это последовательно:

```python
plt.scatter(2, 3)
plt.scatter(4, 7)
plt.scatter(5, 2)
plt.show()
```

Аналогичную визуализацию можно выполнить и с помощью одного оператора scatter:

```python
plt.scatter([2, 4, 5], [3, 7, 2])
plt.show()
```

В вышеуказанном фрагменте кода сначала перечисляются все первые координаты (в виде списка в порядке следования), затем все вторые координаты.
